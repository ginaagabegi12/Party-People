```python
import pandas as pd
import matplotlib.pyplot as plt

#load dataset in
block_party_data = pd.read_csv('block-party.csv')

#dataset structure
print(block_party_data.head())

#count the amount of block parties at a specfic address
address_counts = block_party_data['Address'].value_counts().reset_index()
address_counts.columns = ['Address', 'Block Party Count']

#top 10 addresses by block party count into a graph
top_addresses = address_counts.nlargest(10, 'Block Party Count')

plt.figure(figsize=(10, 6))
plt.barh(top_addresses['Address'], top_addresses['Block Party Count'])
plt.xlabel('Number of Block Parties')
plt.ylabel('Address')
plt.title('Top Addresses for Community-Based Fun in Pittsburgh')
plt.gca().invert_yaxis()
plt.show()

```
