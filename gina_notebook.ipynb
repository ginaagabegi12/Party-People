```python
# Import pandas
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

# Load the dataset
block_party_data = pd.read_csv('block-party.csv')  

# Show the first 10 rows of the data
print(block_party_data.head(10))

# Count the number of block parties at each address
address_counts = block_party_data['Address'].value_counts()  # Count events by address
address_counts = address_counts.reset_index()  # Reset index to make it a DataFrame
address_counts.columns = ['Address', 'Block Party Count']  # Rename columns for clarity

# Select the top 10 addresses with the most block parties
top_addresses = address_counts.head(10)

# Plot the results
plt.figure(figsize=(10, 6))
plt.barh(top_addresses['Address'], top_addresses['Block Party Count'])
plt.xlabel('Number of Block Parties')
plt.ylabel('Address')
plt.title('Top Addresses for Block Parties in Pittsburgh')
plt.show()

```
